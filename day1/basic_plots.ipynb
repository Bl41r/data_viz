```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0, 5, 0.1)
y = np.exp(x)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(1,2,1) # num rows, num columns then which cell to use
plt.scatter(x, y, s=2, edgecolor='None')  # s is size o dots, default is 20
plt.xlabel('Some stuff')
plt.ylabel('Some junk')
plt.title('title1')
plt.suptitle('overall title')
# plt.figure(figsize=(5, 5))
plt.fill_between(x, np.zeros(len(x)), np.cos(x))
plt.ylim(0.5, -0.5) # limit y values of plot
plt.xticks(np.linspace(0,5,6)) # set ticks

plt.subplot(2,2,2)
plt.plot(x,y,color='red')
plt.fill_between(x, y)

plt.show()
```

scatter will make the points all objects, but plot will treat it all as one object
