```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
loans = pd.read_csv('prosperLoanData_truncated.csv')

loans.columns
```

```python
plt.hist(loans.StatedMonthlyIncome)
plt.show()
```

```python
loans.StatedMonthlyIncome.describe()
```

```python
sum(loans.StatedMonthlyIncome > 1E5)
```

```python
plt.figure(figsize=(10, 3))
plt.hist(loans.StatedMonthlyIncome[(loans.StatedMonthlyIncome > 500) & (loans.StatedMonthlyIncome < 40000)], bins=np.arange(0, 40000, 1000))
plt.xlabel('Stated Monthly Income')
plt.ylabel('Number of borrowers')
plt.show()
```

```python
# frequency distribution
heights, edges = np.histogram(loans.StatedMonthlyIncome[(loans.StatedMonthlyIncome > 500) & (loans.StatedMonthlyIncome < 40000)], bins=np.arange(0, 40000, 1000))
probs = heights / sum(heights)
plt.bar(edges[:-1], probs, width=edges[1] - edges[0])
plt.show()
```
